```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn import linear_model
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM,  Flatten, Dropout
from tensorflow.keras.optimizers import Adam
```

```python
test = pd.read_csv("test.csv", encoding="utf-8", index_col=0)
```

```python
x1 = test.values[:,:101]
y = test.values[:, 101:]
```

```python
sample_submission = pd.read_csv("sample_submission.csv", encoding="utf-8", index_col=0)
```

```python
x= x1.reshape(x1.shape[0], 1, x1.shape[1])
```

```python
x.shape
```

```python
model = Sequential()
model.add(LSTM(128, input_shape = (1, 101), return_sequences=True))
model.add(LSTM(128,  return_sequences=False))

model.add(Dense(128, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(64, activation = "relu"))
model.add(Dropout(0.5))
model.add(Dense(12, activation = "linear"))
```

```python
losses=tf.compat.v1.losses.log_loss
model.compile(loss = 'mse', optimizer='adam') 
```

```python
hist_model = model.fit(x,y,
                       epochs = 100,
                       batch_size = 256,
                       verbose = 2, 
                       validation_split=0.2)
```

```python
x_test = test.values[:,:101]
x_test = x_test.reshape(x_test.shape[0], 1, x_test.shape[1])
pred = model.predict(x_test)
submission = pd.DataFrame(data=pred, columns=sample_submission.columns, index=sample_submission.index)
submission.to_csv('submission.csv')
```
